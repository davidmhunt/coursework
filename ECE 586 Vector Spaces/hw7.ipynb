# Problem 1

```python
import numpy as np

P1 = np.array([[1,2,3,4]]).T
P2 = np.array([[4,-2,-6,-7]]).T
P3 = np.array([[3,4,-2,1]]).T

A = np.array([
    [1,4,3],
    [2,-2,4],
    [3,-6,-2],
    [4,-7,1]
])

x = np.array([[1,2,3,7]]).T

# compute projection matrix
G = np.matmul(A.T,A)
G_inv = np.linalg.inv(G)
P = np.matmul(G_inv,A.T)
P = np.matmul(A,P)

x_hat = np.matmul(P,x)
e = x - x_hat

out_str = "P:\n {}\nx_hat: \n{}\ne:\n{}\n".format(P,x_hat,e)
print(out_str)
```

# Problem 3

```python
import numpy as np
import matplotlib.pyplot as plt

G = np.array([
    [1/5,1/4,1/3],
    [1/4,1/3,1/2],
    [1/3,1/2,1]
])
e_1 = np.exp(1)
t = np.array([[e_1 - 2,1,e_1 -1]]).T

c = np.matmul(np.linalg.inv(G),t)
coeffs = c[:,0]

x = np.arange(0,1,0.05)
y_exp = np.array([np.exp(i) for i in x])
y_poly = np.array([np.polyval(coeffs,i) for i in x])
y_diff = np.subtract(y_exp,y_poly)

#plt.plot(x,y_exp)
#plt.plot(x,y_poly)
plt.plot(x,y_diff)


```
