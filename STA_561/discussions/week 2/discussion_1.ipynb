```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LinReg
```

```python
X = np.linspace(0,10,1000)
eps = rand.normal(loc=0,scale=10,size=1000)
y = (1.245 * X**2 + eps)

X = X.reshape(-1,1)
y = y.reshape(-1,1)

model=LinReg()
model.fit(X,y)
y_pred = model.predict(X)

plt.scatter(X,y,label='synthetic data')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(X,y_pred,color='red',label='prediction')
plt.title('Python Linear Regression Demo, coef: %.2f, intercept: %.2f' %(model.coef_,model.intercept_))
plt.legend()
plt.show()
```
