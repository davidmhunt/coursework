```python
def plot_dist(theta,p_theta,label,title="Posterior Distribution",show=True):
    """Plots the given posterior distribution

    Args:
        theta (np.array): N x 1 array of theta values
        p_theta (np.array): N x 1 array of posterior distribution probabilities corresponding to the theta values
        label (_type_): label to appear in the legend
        title (str, optional): Title of the plot. Defaults to "Posterior Distribution".
        show (bool, optional): Displays the plot on True. Defaults to True.
    """
    plt.plot(theta,p_theta,label=label)
    plt.xlabel("theta")
    plt.ylabel("P(Theta|x)")
    plt.title(title)

    if show:
        plt.legend()
        plt.show()
```

```python
#define a function to compute the posterior distribution
def p_6_posterior(theta,x,epsilon):

    A = (epsilon / np.sqrt(2 * np.pi) * np.exp(-0.5 * np.power(theta - x,2)))
    B =  (1 - epsilon)/(np.sqrt(2 * np.pi)) * np.exp(-0.5 * np.power(theta + x,2))

    return  A + B


theta = np.arange(-5,5,0.1)
x = 3
#compute for Epsilon = 1/2
Epsilon = 0.5
p_theta = p_6_posterior(theta,x,Epsilon)

plot_dist(theta,p_theta,"Epsilon = 0.5",title="Posterior Distribution (X = 3)",show=False)

#compute for Epsilon = 3/4
Epsilon = 0.75
p_theta = p_6_posterior(theta,x,Epsilon)

plot_dist(theta,p_theta,"Epsilon = 0.75",title="Posterior Distribution (X = 3)",show=True)
```

## Problem 7

```python
def p_7_posterior(theta,x):

    p_theta = np.exp(-1 * (theta - x))
    p_theta[theta < x] = 0
    return  p_theta

theta = np.arange(0,10,0.1)
x = 3
p_theta = p_7_posterior(theta,x)

plot_dist(theta,p_theta,"","Problem 7 Posterior Distribution (X = 3)")
```
